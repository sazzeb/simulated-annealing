#### **Create a function to remove stop words and convert all string to lower case**

```python
def punctuate_remove_convert_lower(text):
  word_list = []
  punctuations_char = "!\"#$%&'()*+,-./:;<=>?@[\\]^_`{|}~"
  for word in text:
    char = re.sub(r'\S*@\S*', '', word)
    char = char.lower().translate(str.maketrans('', '', punctuations_char))
    word_list.append(char)
  return word_list
```
